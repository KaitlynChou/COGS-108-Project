# Abstract

For our COGS 108 group project, we sought to find a correlation between children’s participation at their local county libraries with their test performance on the California Assessment of Student Performance and Progress (CAASPP) from grades 3 through 8. To further define what children’s participation entails, we focused on the circulation of children’s materials, the number of child borrowers, the number of child programs, and the program attendance at libraries. Initially, we predicted that all four of these factors were positively correlated with the percentage of standard exceeded in both ELA and mathematics CAASPP tests. However, after diving deeper into our datasets and conducting analyses, we discovered that only one of the variables had the most significant positive correlation with the percentage of standard exceeded on the CAASPP tests – the circulation of children’s materials in California county libraries. Program attendance also had a moderate correlation, but it was not as significant as the circulation of children’s materials.

# Research Question

Does participation (number of registered users, amount of content circulated, programs attended) at libraries in California counties correlate to CAASPP test scores (ELA and Mathematics) of students attending school in those counties?

## Data Overview

### Dataset #1: [California Public Libraries Statistics](https://www.library.ca.gov/stats/?pageNo=1)

This public database provides "annual statistics provides annual statistics about libraries’ finances, services, programs, and collections." This data is collected by state libraries through the The California Public Libraries Survey, and the data is reported to the Institute of Museum and Library Services. The oldest data available currently is from 2016-17. We will be using data from 2016-2017 to 2021-2022, excluding 2019-2020. This is because we are missing state testing data for 2020 and thus cannot compare library participation to testing data for that year.

Notable variables: 

**`Children Borrowers`**: Number of children registered with the library with a child/youth library card. Specific age range for having a child/youth card can vary by system.

**`Library Visits`**: Number of persons who entered the library during the year. Can be estimated by counting visits for a typical week in October, times 52.

**`Circulation of Childrens Materials`**: Total number of times all materials (books, electronic materials, DVDs, CDs, etc.) marked as Children's were checked out or renewed during the year. Includes checkouts by everyone, including those with regular library cards instead of child/youth cards.

**`# of Children's Programs`**: Number of events planned by the library aimed for children ages 6-11, often to provide cultural, recreational, or educational information. 

**`Children's Program Attendance`**: Sum of number of attendees at all children's programs, whether in-person or virtual

### Dataset #2: [California Assessment of Student Performance and Progress (CAASPP) Smarter Balanced Assessments](https://caaspp-elpac.ets.org/caaspp/ResearchFileListSB?ps=true&lstTestYear=2023&lstTestType=B&lstCounty=00&lstDistrict=00000&lstFocus=a)

Every year (typically between January and June of a school year), students grades 3-8 and 11 in California take the California Assessment of Student Performance and Progress Smarter Balance assessments in English Language Arts/Literacy (ELA) and Mathematics. These tests serve to measure students' learning based on Common Core State Standards. Each student receives a Scale Score and is categorized into one of four levels based on their scale score: Standard Exceeded, Standard Met, Standard Nearly Met, and Standard Not Met. The calculations for what score belongs in which category can be found [here](https://caaspp-elpac.ets.org/caaspp/ScaleScoreRangesSB), and more in-depth definitions of the standard levels are found [here.](https://caaspp-elpac.ets.org/caaspp/UnderstandingSBResults#am1) The public research files online contain mean scale scores as well as the percentages of students who fall in each category, from the county level down to the school level. We will be using county level data. We will also only be focusing on grades 3 through 8, because grade 11 falls out of the Child Borrower range for most public libraries. The research files date back to the 2014-15 school year, but we will use data beginning from the 2016-17 school year to match the library data that we have.

Notable variables:

**`County Code`**: Number between 1 and 58 (inclusive) that represents a county. Will be replaced with County Name using a separate dataset provided by CAASPP that matches county names to the codes.

**`Test ID`**: Test ID = 1 corresponds to the ELA test. Test ID = 2 corresponds to the Math test. We will make these separate columns and then delete this column.

**`Percentage Standard Exceeded`**: Percentage of tested students in a grade whose scale scores exceed the standard. 

**`Percentage Standard Met`**: Percentage of tested students in a grade whose scale scores meet the standard.

**`Percentage Standard Nearly Met`**: Percentage of tested students in a grade whose scale scores nearly meet the standard.

**`Percentage Standard Not Met`**: Percentage of tested students in a grade whose scale scores do not meet the standard.

For the percentage columns, we will take the mean of all grades 3-8 across each county to get the county's mean percentages. We will use these values in our county-level analysis. The possible ranges for scale score varies for each grade, so we will focus on using the percentages in each category to evaluate the overall performance of students in each county.

### Dataset #3: California County Population by Characteristics: [2010-2019](https://www.census.gov/data/datasets/time-series/demo/popest/2010s-counties-detail.html) & [2020-2022](https://www.census.gov/data/datasets/time-series/demo/popest/2020s-counties-detail.html)

The United States Census Bureau's Population Estimates Program publishes estimates for state populations based on the rates of births, deaths, and migrations in a given year. This dataset estimates the population of California counties, separated by age groups and sex. We are only interested in the number of children aged 5-13 as this age group most closely matches the population who would own children's/youth library cards and are taking the CAASPP Smarter Balance tests. We are using data from 2017, 2018, 2019, 2021, and 2022. We must retrieve 2017-2019 and 2021-22 separately because they are in two different csv files. Using the population, we will calculate library participation **per 1,000 children** in a particular county, rather than using the raw numbers.

### #1 Cleaning California Public Libraries Statistics

We then combined all five years into one DataFrame.

```python
# Read in csv 
library_17 = pd.read_csv('./data/All-DataFY16-17.csv',  encoding='latin-1')

# Select only columns we are interested in 
lib_var = ['County', 'Population of The Legal Service Area', 
        'Registered Users as of June 30', 'Children Borrowers', 
        'Library Visits', 'Total Operating Income', 'Circulation of Childrens Materials',
        "# of Children's Programs", "Children's Program Attendance"]
library_17 = library_17[lib_var]

# Convert quantitative variables to floats 
library_17[lib_var[1:]] = library_17[lib_var[1:]].apply(lambda x: x.str.replace('$','')).apply(lambda x: x.str.replace(',','')).astype(np.float64)
library_17['Year'] = 2017

# Do the same with other years
library_18 = pd.read_csv('./data/All-DataFY17-18.csv',  encoding='latin-1')
library_18 = library_18[lib_var]
library_18[lib_var[1:]] = library_18[lib_var[1:]].apply(lambda x: x.str.replace('$','')).apply(lambda x: x.str.replace(',','')).astype(np.float64)
library_18['Year'] = 2018

library_19 = pd.read_csv('./data/All-DataFY18-19.csv',  encoding='latin-1')
library_19 = library_19[lib_var]
library_19[lib_var[1:]] = library_19[lib_var[1:]].apply(lambda x: x.str.replace('$','')).apply(lambda x: x.str.replace(',','')).astype(np.float64)
library_19['Year'] = 2019

# Since CAASPP is missing 2020 data, we cannot compare it with 2020 library data

library_21 = pd.read_csv('./data/All-DataFY20-21.csv',  encoding='latin-1')
# Column names differ slightly from previous years, manually adjusted
library_21 = library_21[['1.35 County', '2.1 Population of The Legal Service Area', 
        '2.2 Registered Users as of June 30', '2.3 Children Borrowers', '7.2 Library Visits', 
        '3.5 Total Operating Income', '7.12 Circulation of Childrens Materials', 
        "# of Children's Programs (calculated)", "Children's Program Attendance"]]
# Rename columns 
library_21.columns = lib_var
library_21[lib_var[1:]] = library_21[lib_var[1:]].apply(lambda x: x.str.replace('$','')).apply(lambda x: x.str.replace(',','')).astype(np.float64)
library_21['Year'] = 2021

library_22 = pd.read_csv('./data/All-DataFY21-22.csv',  encoding='latin-1')
library_22 = library_22[['1.36 County', '2.1 Population of The Legal Service Area', 
        '2.2 Registered Users as of June 30', '2.3 Children Borrowers', '7.2 Library Visits', 
        '3.5 Total Operating Income', "7.11 Circulation of Children's Materials", 
        "# of Children's Programs (calculated)", "Children's Program Attendance"]]
library_22.columns = lib_var
library_22[lib_var[1:]] = library_22[lib_var[1:]].apply(lambda x: x.str.replace('$','')).apply(lambda x: x.str.replace(',','')).astype(np.float64)
library_22['Year'] = 2022


libraries = pd.concat([library_17, library_18, library_19, library_21, library_22])
```

### #3 Cleaning California Population Dataset
In the population dataset, we select the County Name, Year, and 'AGE313_TOT' column, which represents the total estimated number of children aged 5-13 in a particular county. The values from 2017-2019 are in a separate file from the values for 2021-22, so we combine them into one dataframe.

```python
# Read in csv for population estimates 2010-2019
pop_17_19 = pd.read_csv('./data/cc-est2019-agesex-06.csv')

# Select years 2017, 2018, 2019
pop_17_19 = pop_17_19[pop_17_19['YEAR'] > 9]

# Select columns county name, year, and population ages 5-13
pop_17_19 = pop_17_19[['CTYNAME', 'YEAR', 'AGE513_TOT', 'POPESTIMATE']]
pop_17_19['YEAR'] = pop_17_19['YEAR'].replace({10: 2017, 11: 2018, 12: 2019})

# Read in csv for population estimates 2020-2022
pop_21_22 = pd.read_csv('./data/cc-est2022-agesex-06.csv')

# Select years 2021 and 2022
pop_21_22 = pop_21_22[pop_21_22['YEAR'] > 2]
pop_21_22 = pop_21_22[['CTYNAME', 'YEAR', 'AGE513_TOT', 'POPESTIMATE']]
pop_21_22['YEAR'] = pop_21_22['YEAR'].replace({3: 2021, 4: 2022})

# Combine the years
population = pd.concat([pop_17_19, pop_21_22])
population = population.sort_values(by=['CTYNAME', 'YEAR'])
```

In the following cell, we are calculating the proportion of Child Borrowers, Circulation of Children's Materials, # of Children's Programs, and Children's Program Attendance **per 1,000 children**. Because counties with more people/children and more libraries will naturally have larger numbers, getting the values proportional to the counties' populations ensures that we are measuring the rate of activity fairly. We will use **libraries_per** to perform our analysis from now on.

```python
# Create a new dataframe for library participation per 1,000 people or per 1,000 children
libraries_per = new_libraries.copy()
del new_libraries

# Get values per 1,000 people/children
libraries_per['Registered Users as of June 30'] = libraries_per['Registered Users as of June 30'] / libraries_per['Population of The Legal Service Area'] *1000
libraries_per['Library Visits'] = libraries_per['Library Visits'] / libraries_per['Population of The Legal Service Area'] *1000
libraries_per['Children Borrowers'] = libraries_per['Children Borrowers'] / libraries_per['Child Population'] *1000
libraries_per['Circulation of Childrens Materials'] = libraries_per['Circulation of Childrens Materials'] / libraries_per['Child Population'] *1000
libraries_per["# of Children's Programs"] = libraries_per["# of Children's Programs"] / libraries_per['Child Population'] *1000
libraries_per["Children's Program Attendance"] = libraries_per["Children's Program Attendance"] / libraries_per['Child Population'] *1000
libraries_per.columns = ['County', 'Year', 'Population of The Legal Service Area', 'Registered Users Per 1k People', 'Children Borrowers Per 1k Children', 
                         'Library Visits Per 1k People', 'Total Operating Income', 'Circ. Childrens Materials Per 1k Children', "Children's Programs Per 1k Children",
                       "Children's Program Attendance Per 1k Children", 'Child Population']
libraries_per
```

Or do older students' test performance correlation more strongly? If there is a difference, it could be indicative of which grades have more interaction with public libraries.

In this section, we compare the library activity (circulation of materials, #child borrowers, # child programs, child program attendance) in a particular year to the proportions of students who exceeded, met, nearly met, or didn't meet the state standard.

```python
# Create new dataframes for every year's library activity and test performance to graph correlation
caaspp_dfs = {}
lib_dfs = {}
year_dfs = {}
for i in range (2017, 2023):
    if i==2020: # skip 2020
        continue
    # add the library variables from this year    
    lib_dfs[f"{i}"] = pd.DataFrame({'Year': i, 'County': libraries_per[libraries_per['Year']==i]['County'].reset_index(drop=True),
                        'Circ. Childrens Materials': libraries_per[libraries_per['Year']==i]['Circ. Childrens Materials Per 1k Children'].reset_index(drop=True),
                        'Child Borrowers': libraries_per[libraries_per['Year']==i]['Children Borrowers Per 1k Children'].reset_index(drop=True),
                        'Child Programs': libraries_per[libraries_per['Year']==i]["Children's Programs Per 1k Children"].reset_index(drop=True),
                        'Prgm Attendance': libraries_per[libraries_per['Year']==i]["Children's Program Attendance Per 1k Children"].reset_index(drop=True)})
    
    # add the CAASPP variables from this year
    for j in range (3, 9):
        if j==3:
            caaspp_dfs[f"{i}"] = pd.DataFrame({'County': caaspp[(caaspp['Test Year']==i) & (caaspp['Grade']==j)]['County Name'].reset_index(drop=True),
                                    'Year': caaspp[(caaspp['Test Year']==i) & (caaspp['Grade']==j)]['Test Year'].reset_index(drop=True)})
        caaspp_dfs[f"{i}"][f"%ExceededELA_{j}"] = caaspp[(caaspp['Test Year']==i) & (caaspp['Grade']==j)]['Percent Exceeded ELA'].reset_index(drop=True)
        caaspp_dfs[f"{i}"][f"%ExceededMath_{j}"] = caaspp[(caaspp['Test Year']==i) & (caaspp['Grade']==j)]['Percent Exceeded Math'].reset_index(drop=True)
        caaspp_dfs[f"{i}"][f"%MetELA_{j}"] = caaspp[(caaspp['Test Year']==i) & (caaspp['Grade']==j)]['Percent Met ELA'].reset_index(drop=True)
        caaspp_dfs[f"{i}"][f"%MetMath_{j}"] = caaspp[(caaspp['Test Year']==i) & (caaspp['Grade']==j)]['Percent Met Math'].reset_index(drop=True)
        caaspp_dfs[f"{i}"][f"%NearELA_{j}"] = caaspp[(caaspp['Test Year']==i) & (caaspp['Grade']==j)]['Percent Nearly Met ELA'].reset_index(drop=True)
        caaspp_dfs[f"{i}"][f"%NearMath_{j}"] = caaspp[(caaspp['Test Year']==i) & (caaspp['Grade']==j)]['Percent Nearly Met Math'].reset_index(drop=True)
        caaspp_dfs[f"{i}"][f"%NotELA_{j}"] = caaspp[(caaspp['Test Year']==i) & (caaspp['Grade']==j)]['Percent Not Met ELA'].reset_index(drop=True)
        caaspp_dfs[f"{i}"][f"%NotMath_{j}"] = caaspp[(caaspp['Test Year']==i) & (caaspp['Grade']==j)]['Percent Not Met Math'].reset_index(drop=True)
    
    # merge the library variables and the CAASPP variables into one dataframe for this year
    year_dfs[f"{i}"] = pd.merge(lib_dfs[f"{i}"], caaspp_dfs[f"{i}"], left_on=['Year', 'County'], right_on=['Year', 'County'])

df_2017 = year_dfs['2017']
df_2018 = year_dfs['2018']
df_2019 = year_dfs['2019']
df_2021 = year_dfs['2021']
df_2022 = year_dfs['2022']
del year_dfs, caaspp_dfs, lib_dfs
df_2017.head()
```

We are interested in how these correlations might change with time. Rather than plotting the above pairplot for each year and each test (ELA and math), let's define a function that gives us the variables with the strongest correlations, using the Pearson correlation coefficient. We are also going to see if there are correlations for the other Standard categories (Standard Met, Nearly Met, and Not Met).

```python
# Function that retrieves the variable pairs with the highest correlation values for each year
math_all = ['%ExceededMath_3', '%ExceededMath_4', '%ExceededMath_5', '%ExceededMath_6', '%ExceededMath_7', '%ExceededMath_8',
         '%MetMath_3', '%MetMath_4', '%MetMath_5', '%MetMath_6', '%MetMath_7', '%MetMath_8',
         '%NearMath_3', '%NearMath_4', '%NearMath_5', '%NearMath_6', '%NearMath_7', '%NearMath_8',
         '%NotMath_3', '%NotMath_4', '%NotMath_5', '%NotMath_6', '%NotMath_7', '%NotMath_8']
    
ela_all = ['%ExceededELA_3', '%ExceededELA_4', '%ExceededELA_5', '%ExceededELA_6', '%ExceededELA_7', '%ExceededELA_8',
         '%MetELA_3', '%MetELA_4', '%MetELA_5', '%MetELA_6', '%MetELA_7', '%MetELA_8',
         '%NearELA_3', '%NearELA_4', '%NearELA_5', '%NearELA_6', '%NearELA_7', '%NearELA_8',
         '%NotELA_3', '%NotELA_4', '%NotELA_5', '%NotELA_6', '%NotELA_7', '%NotELA_8']

def pos_corr(df, year):
    # Make a correlation coefficient matrix for all variables of interest
    corr = df[lib_vars + ela_all + math_all].corr()
    # Only include unique x-axis and y-axis variables
    corr = corr[ela_all + math_all]
    corr = corr.loc[lib_vars]
    
    # Display the variable pairs with the strongest correlations
    pos_corr_df = corr.unstack().nlargest(5).reset_index()
    pos_corr_df.columns = ['CAASPP Variable', 'Library Variable', 'Strongest Pos. Correlation']
    pos_corr_df['Year'] = year
    pos_corr_df = pos_corr_df[['Year', 'CAASPP Variable', 'Library Variable', 'Strongest Pos. Correlation']]
    return pos_corr_df

# Get the strongest correlation coefficients from each year
strong_pos_corr = pd.concat([pos_corr(df_2017, 2017), pos_corr(df_2018, 2018), pos_corr(df_2019, 2019),
                               pos_corr(df_2021, 2021), pos_corr(df_2022, 2022)])
strong_pos_corr
```

The top five strongest correlations in each year are compiled in the dataframe strong_pos_corr above. Across all years, there are consistently correlation coefficients greater than 0.8 (indicating a **strong positive correlation**) between the **circulation of children's materials** in a county and the percentage of students who **exceeded** the state standard in that county, in both the ELA and Math tests. This aligns with what we saw on the pairplots, where the most linear distribution of points was between circulation of materials and percentage standard exceeded. These results mean that counties with higher rates of chilren's material circulation see a larger percentage of students who performed very well on the state test. 

There is also some correlation between Program Attendance and percentage exceeded, which we saw in the 2017 pairplots. However, the correlation is not very strong with a coefficient of 0.502. 

```python
# Which grades and subjects have the most correlations with library material circulation?
strong_pos_corr.value_counts('CAASPP Variable')
```

There is a wide distribution of grades that have the strongest correlations, which leads us to believe there is not one particular age group or grade whose performance has a stronger correlation to library material circulation than the others. However, the value_counts() method above shows us that grades 3 and 4 have the most strong correlations with Circulation of Children's Materials, particularly on the Math test.

In the next cell, we repeat this process to test for negative correlations.

```python
def neg_corr(df, year):
    # Make a correlation coefficient matrix for all variables of interest
    corr = df[lib_vars + ela_all + math_all].corr()
    # Only include unique x-axis and y-axis variables
    corr = corr[ela_all + math_all]
    corr = corr.loc[lib_vars]
    
    # Display the variable pairs with the 10 strongest correlations
    neg_corr_df = corr.unstack().nsmallest(5).reset_index()
    neg_corr_df.columns = ['CAASPP Variable', 'Library Variable', 'Strongest Neg. Correlation']
    neg_corr_df['Year'] = year
    neg_corr_df = neg_corr_df[['Year', 'CAASPP Variable', 'Library Variable', 'Strongest Neg. Correlation']]
    return neg_corr_df

strong_neg_corr = pd.concat([neg_corr(df_2017, 2017), neg_corr(df_2018, 2018), neg_corr(df_2019, 2019),
                               neg_corr(df_2021, 2021), neg_corr(df_2022, 2022)])
strong_neg_corr
```

There is definitely some **negative correlation** between the **circulation of children's materials** in a county and the percentage of students who **nearly met** or **did not meet** the state standards, though not as strong as the positive correlations. While the positive correlation coefficients reached up to 0.83, the strongest negative correlation is -0.74 and most are between 0.5 and 0.7, suggesting moderate negative correlation overall. 

The results from the strong_neg_corr dataframe mean that counties with a higher circulation of children's materials also generally saw a lower percentage of students falling below the state standards in both Math and ELA. Again, the results do not indicate a difference in correlation strength between different grades or age groups. All grades show correlation across various years.

The only CAASPP variable we did not see strong correlation with is the percentage of students who **met** the standard. 

# Ethics & Privacy

Given our research question, one of the primary concerns we have regards the privacy of the children whose data we will be collecting. The data we wish to utilize will track participation at public libraries as well as student test scores. We understand that while parents or guardians may have consented to allow the collection of their children's data, it is unlikely that the children themselves can offer truly informed consent to have their data collected and used. For this reason, we propose to utilize only data that refrains from including all personal identifying information except for the age of children. Additionally, in our analysis, we aim to only use this data in aggregate — never singling out a single data point for commentary on that individual. Another privacy concern hinges on the fact that some of our data includes attendance rates for children's programs at public libraries. To protect the privacy and safety of the children who attend these programs, we will withhold information about the times and locations of the programs as well as consider only using the program data in aggregate by county. Lastly, considering that these libraries are public institutions, the data must be safeguarded through strong security measures to prevent hacking, data leaks, and other security breaches.

The ethics concerns we have for our data, however, extend beyond just privacy. One issue that might arise from the data is the stigmatization of certain communities or school systems. Since we include performance data on standardized tests as well as library attendance habits, viewers may interpret these as metrics to profile certain groups or communities. In addition, inherent biases can impact the validity and fairness of research findings, which may result in overrepresentation or underrepresentation of particular demographics and socioeconomic groups. Lastly, when collecting data within communities, it is imperative to implement responsible measures and practices that leave the community feeling safe and at ease. It is important in our analysis to remind readers that standardized test scores are not a holistic measure of an individual's capabilities nor is library attendance an accurate means to assess a community's habits.

Our preconceptions about the likely academic benefit of library engagement among young people is supported by a positive correlation between library participation and success on CAASPP.  One can point to the myriad of ways by which libraries serve as invaluable resources for students throughout their education. Because such library engagement supports early habits of reading and academic focus time, we expect that children who participate at their local libraries more frequently will have better performance on the CAASPP tests. The findings of our project solidify these ideas with quantitative measurements of how standardized academic excellence can be predicted by library participation.

There are a few limitations to our approach, one being the consideration that libraries may collect their data differently. The variables ‘Circulation of Children’s Materials’ and ‘Program Attendance’, for instance, could suggest different things depending on how they are collected and considered across libraries. Another potential limitation is that not all of the children who participate in library activities, such as borrowing books or attending children’s programs, take the California Assessment of Student Performance and Progress (CAASPP). Although the test is mandated across all public schools in California, children who attend private schools or are homeschooled are not required to take the CAASPP, making our results difficult to generalize across all children who attend and participate at their local libraries. We must keep these limitations in mind when considering how generalizable our findings are.

Moving forward, these findings could be utilized to warrant researching the causality between library participation and CAASPP standardized test scores. One might approach this task by selecting groups of children distributed randomly and evenly across all California counties and encouraging – perhaps parentally - differing levels of library engagement from each group over a predetermined period of time. Measuring the students' CAASPP scores after this period may provide insight into whether or not library participation is causal in relation to the scores. While this proposed research plan stands as a rough approach, much oversight and imagination would be needed to refine such a study to ensure that any ethical concerns are alleviated. Especially when dealing with children, there is only so much that can be done to control an experiment while maintaining a firm level of ethicality. However, even without taking further steps to establish potential causality, our findings can still be used by state government and education officials as a baseline to encourage young people and their parents to develop library engagement habits that will position the students for stronger academic success.
